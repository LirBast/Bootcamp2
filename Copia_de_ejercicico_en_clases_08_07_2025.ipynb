```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(55)   ###siempre me van a dar los mismos resultados con esta seed

data={'Edad':np.random.randint(18,25,100),
       'Horas de estudio':np.random.randint(5,30,100),
       'Calificacion':np.random.normal(75,10,100).clip(0,100),
       'Genero':np.random.choice(['Femenino','Masculino'],100),
       'Deporte':np.random.choice(['Futbol','Baloncesto','Natacion','Volleyball'],100)}     ### .clip(0,100) fuerza a que los registros siempre se encuentren entre 0 a 100 por los datos que se ingresaron

df=pd.DataFrame(data)
print(df.head())

sns.histplot(data=df,x='Calificacion',kde=True,bins=20,color='green') ### el kde aade la curva de densidad
plt.title('Histograma de Calificaciones')
plt.xlabel('Calificacion')
plt.ylabel('Frecuencia')
plt.show()

sns.kdeplot(data=df,x='Horas de estudio',shade=True,color='red')   ### shade rellena los valores de bajo de la curva
plt.title('Estimacion de densidad: Horas de estudio')
plt.xlabel('Horas de estudio')
plt.ylabel('Densidad')
plt.show()

sns.histplot(data=df,x='Horas de estudio',hue='Genero',kde=True,bins=20,color='purple',multiple='dodge')
plt.title('Diostribucion de horas de estudio por genero')
plt.xlabel('Horas de estudio')
plt.ylabel('Densidad')
plt.legend(title='Genero')
plt.show()

sns.jointplot(data=df,x='Horas de estudio',y='Calificacion',color='grey',kind='kde')
plt.suptitle('Relacion Horas de estudio y Calificacion',y=1.02)
plt.show()

sns.pairplot(df[['Edad','Horas de estudio','Calificacion','Genero']],hue='Genero',diag_kind='kde')
plt.suptitle('Matriz de edad, horas de estudio y calificaciones',y=1.02)
plt.show()


```
