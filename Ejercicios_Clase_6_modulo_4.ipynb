```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(44)


meses=['junio','julio','agosto','septiembre']
venta=[150,100,120,90]

### grafico de lineas

plt.plot(meses,venta,marker='.')
plt.title('Venta de helados')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.show()

### Fiuras y subplot

fig, ax=plt.subplots(figsize=(8,5))   ###  el primer es el ancho y el segundo es el alto
ax.plot(meses,venta,color='green',linestyle=':',marker='o')
ax.set_title('Venta de helado (con ax)')
ax.set_xlabel('meses')
ax.set_ylabel('ventas')
ax.grid(True)
plt.xticks(rotation=45,ha='right')
ax.legend(['ventas_mensuales'])
plt.tight_layout()
plt.show()

tipos_helados=['chocolate','vainilla','pistacho','menta']
venta_totales=[250,200,350,100]

fig, ax=plt.subplots(figsize=(8,6))
ax.bar(tipos_helados,venta_totales,color=['chocolate','yellow','lightgreen','green'])
ax.set_title('Ventas totales vs venta de helado')
ax.set_xlabel('tipo de helado')
ax.set_ylabel('ventas totales')
plt.xticks(rotation=30,ha='right')
ax.grid(True)
plt.tight_layout()
plt.show()


### histograma

precio_helados=np.random.normal(loc=3.5,scale=0.5,size=200)

fig, ax=plt.subplots(figsize=(7,5))
ax.hist(precio_helados,bins=10,edgecolor='black',alpha=0.7)  #alpha se usa para transparencia de las barras del histograma
plt.show()


temperatura=np.linspace(20,35,30)
ventas_diarias=50+temperatura*3+np.random.normal(0,10,30)

fig, ax=plt.subplots(figsize=(7,5))
ax.scatter(temperatura,ventas_diarias,color='cyan',alpha=0.8,edgecolor='black')
ax.set_xlabel('temperatura en grados centigrados')
ax.set_ylabel('ventas diarias')
ax.grid(True, linestyle=':', alpha=0.6)
plt.show()


### diagrama de caja y bigote

ventas_sucursal_a=np.random.normal(150,30,100)
ventas_sucursal_b=np.random.normal(120,40,100)
ventas_sucursal_b=np.append(ventas_sucursal_b,[50,250,260]) #total de datos b:103
datos=[ventas_sucursal_a,ventas_sucursal_b]
nomnbre_sucursales=['sucursal A','Sucursal B']


fig, ax=plt.subplots(figsize=(7,5))
ax.boxplot(datos,labels=nomnbre_sucursales,medianprops={'color':'red'},patch_artist=True)
ax.set_title("Distribucion de ventas diarias por sucursal")
ax.set_xlabel('sucursales')
ax.set_ylabel('ventas diarias')
ax.grid(True)
plt.show()


### ejercicio 2 variables

ventas_sucursal_a=np.random.normal(150,30,100)
ventas_sucursal_b=np.random.normal(120,40,100)
ventas_sucursal_b=np.append(ventas_sucursal_b,[50,250,260]) #total de datos b:103
datos=[ventas_sucursal_a,ventas_sucursal_b]
nomnbre_sucursales=['sucursal A','Sucursal B']
ventas_vendedor_1=[80,130,100,70]
ventas_vendedor_2=[90,140,110,80]

fig, ax=plt.subplots(figsize=(8,5))
ax.plot(meses,ventas_vendedor_1,label='vendedor 1', color='purple',marker='o')
ax.plot(meses,ventas_vendedor_2,label='vendedor 2', color='grey',marker='o')
ax.set_title("Comparacion ventas mensuales por vendedor")
ax.set_xlabel('meses')
ax.set_ylabel('ventas')
ax.legend()
ax.grid(True)
plt.tight_layout()
fig.savefig('Venta_helados.png', dpi=300, bbox_inches='tight')
plt.show()



```
