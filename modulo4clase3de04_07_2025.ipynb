```python
import numpy as np
import pandas as pd


# np.random.seed(777)   ### si quiero tener siempre los mismo numeros le aplico la seed

genero=np.random.choice(['Hombre','Mujer'],size=50)

preferencias=[]

for g in genero:
    if g == 'Mujer':
        preferencias.append(np.random.choice(['si','no'],p=[0.6,0.4]))
    else:  # Hombre
        preferencias.append(np.random.choice(['si','no'],p=[0.4,0.6]))

datos = pd.DataFrame({'Genero': genero,
                      'Preferencia': preferencias})

print(datos.head())



## tabla de contingencia

tabla_contingencia=pd.crosstab(datos['Genero'],datos['Preferencia'])

print('----------------------------')
print(tabla_contingencia)
print('----------------------------')
cantidad_mujeres=tabla_contingencia.loc['Mujer'].sum() ### el loc sirve para acceder a fila o columnas del data frame
## Cantidad de mujeres
print(cantidad_mujeres)
print('----------------------------')
# mujeres que si quieren ir al cine
mujeres_pref_cine=tabla_contingencia.loc['Mujer','si'].sum()
print(mujeres_pref_cine)
print('----------------------------')
probabilidad_mujer_si=mujeres_pref_cine/cantidad_mujeres
##probabilidad de mujeres que si quieren ir al cine
print(probabilidad_mujer_si)
print('----------------------------')
### Hombres que no quieren ir al cine
hombres_pref_cine=tabla_contingencia.loc['Hombre','no'].sum()
print(hombres_pref_cine)


```

```python
### Grafico de Diagrama de dispersion
import matplotlib.pyplot as plt
import numpy as np

horas_estudio=np.random.uniform(1,10,20)
nota_examen=horas_estudio*0.8+np.random.normal(0,1,20)

df=pd.DataFrame({'horas_de_estudio':horas_estudio,
                 'Notas_examen':nota_examen})
print(df)

plt.figure(figsize=(8,6))
plt.scatter(df['Notas_examen'],df['horas_de_estudio'])
plt.xlabel('Horas de estudio')
plt.ylabel('Nota en examen')
plt.title('Diagrama de dispersi√≥n: Horas de estudio vs Nota examen')
plt.grid(True)
plt.show()
```
